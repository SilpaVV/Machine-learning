# **Data Pre-processing of the bioactivity data**
**Labelling compounds as ACTIVE,INACTIVE AND INTERMEDIATE**

Bioactivity data is in the Kd unit.
The compounds having **less than 1000 nM** value is considered - **ACTIVE*
The compounds having **greater  than 10000 nM value** is considered-**INACTIVE**
The compounds having values between **1000-10000nM** is considered - **INTERMEDIATE**





```python
bioactivity_class=[]
for i in df2.standard_value:
  if float(i)>=10000:
    bioactivity_class.append("inactive")
  elif float(i)<=1000:
      bioactivity_class.append("active")
  else:
        bioactivity_class.append("intermediate")
```

```python
df2.molecule_chembl_id
```

```python
df2.standard_value
```

```python
selection=['molecule_chembl_id','canonical_smiles','standard_value']
df3=df2[selection]
```

```python
df3.head(4)
```

```python
df4=pd.concat([df3,pd.Series(bioactivity_class)],axis=1)
```

# **Calculate Lipinski Descriptors**
Rule of five to evaluate the drug likeness of compounds based on key pharmacokinetic properties: **ABSORPTION,DISTRIBUTION,METABOLISM,EXTRACTION**
Lipinski's rule stated the following:


*   Molecular weight < 500 Dalton
*   Octanol-water partition coefficient(LogP)<5
*  Hydrogen bond donors
*   Hydrogen bond acceptors<10
All the decriptors are multiples of 5
